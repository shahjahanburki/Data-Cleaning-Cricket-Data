Spliting up columns

```python
df['Span'].str.split(pat= '-')
```

```python
df['start_year'] = df['Span'].str.split(pat= '-').str[0]
df['end_year'] = df['Span'].str.split(pat= '-').str[1]

df.head()
```

```python
df1 = df.copy()
df1
```

```python
df1= df1.drop(['Span'], axis = 1)

df1.head()
```

```python
df1['Player'].str.split(pat='(').str[1]
```

```python
df1['country'] = df1['Player'].str.split(pat='(').str[1]
df1
```

```python
df1['country'] = df1['country'].str.split(pat=')').str[0]
df1['country']
```

```python
df1['Player'] = df1['Player'].str.split(pat='(').str[0]

df1.head()
```

```python
df.dtypes
```

```python
df1['highest_inn_score'] = df1['highest_inn_score'].str.split(pat='*').str[0]
```

```python
df1['highest_inn_score'] = df1['highest_inn_score'].astype('int')

df1.dtypes
```

```python
df1 = df1.astype({'start_year':'int', 'end_year':'int'})
df1.dtypes
```

```python
df1['Mat'] = df1['Mat'].astype('str')
```

```python
df1['Mat'].str.split(pat='*').str[0]
```

```python
df1['balls_faced'] = df1['balls_faced'].str.split('+').str[0]

df1['balls_faced'] = df1['balls_faced'].astype('float')
```

```python
mean_balls_faced = df1['balls_faced'].mean()

df1['balls_faced'] = df1['balls_faced'].fillna(mean_balls_faced)
df1['balls_faced'].isnull().sum()
```

```python
df1['Mat'] = df1['Mat'].astype('int')
```

```python
df1['4s'] = df1['4s'].str.split(pat='+').str[0]

df1['4s'] = df1['4s'].astype('int')
```

```python
df1['6s'] = df1['6s'].str.split(pat='+').str[0]

df1['6s'] = df1['6s'].astype('int')
```

```python
df1.info()
```

```python
df1['career_length'] = df1['end_year'] - df1['start_year']

df1.head()
```

Calculating the average career length

```python
avg_career_length = df1['career_length'].mean()
avg_career_length
```

calculating the avergae striking rate for players who played for over 10 years

```python
df1[df1['career_length']>10]['strike_rate'].mean()
```

Number of cricketers who played before 1960

```python
df1[df1['start_year'] < 1960]['Player'].count()
```

Ranking countries that scored the highest

```python
df1.groupby('country')['highest_inn_score'].max().to_frame('Highest according to country').reset_index().sort_values('Highest according to country')
```

Finding the average 100s, 50s, and 0s scored

```python
df1.groupby('country')[['100', '50', '0']].mean()
```
